```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model =  ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
from pydantic import BaseModel, Field
```

```python
class ResponseStructure(BaseModel):
    name: str = Field(description = "Complete name of the user")
    email : str = Field(description = "Email address")
    age: int = Field(description = "Age of the user")
```

```python
structured_llm = model.with_structured_output(ResponseStructure)
```

```python
text = "Hello, my name is Aman Roy, "\
"Email is aman@gmail.com and my age is 24"
```

```python
response = structured_llm.invoke(f"Please give me only name, email, and age from the text: {text}")
```

```python
response
```

```python
class Movies(BaseModel):
    title: str = Field(description="Movie title")
    year: int = Field(description="Movie release year")
```

```python
movie_llm = model.with_structured_output(Movies)
response = movie_llm.invoke("Give me one recent movie")
print(response)
```
